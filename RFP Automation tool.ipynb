```python
import os
import mimetypes
from docx import Document

filename = input("Enter a filename without extension: ")

# Search for a file with the given name and any extension
for ext in mimetypes.types_map:
    # Combine the filename and extension
    filename_with_ext = f"{filename}{ext}"

    # Check if the file exists
    if os.path.exists(filename_with_ext):
        # Get the MIME type of the file
        mime_type, encoding = mimetypes.guess_type(filename_with_ext)
        # Check if the file is a docx file
        if mime_type == 'application/vnd.openxmlformats-officedocument.wordprocessingml.document':
            print(f"The file {filename_with_ext} is a docx file.")
            for para in Document(filename_with_ext).paragraphs:
                print(para.text)
        else:
            print(f"The file {filename_with_ext} is not a docx file.")
        break
        
else:
    # If no file was found, print an error message
    print("Filepath does not exist. Please enter a valid filepath!")
```

# Table into Bulletin points

```python
import docx

# Open the Word document
doc = Document(filename_with_ext)

# Iterate over all tables in the document
def tables(document):
    for table in doc.tables:

        # Extract the header row from the table
        header = [cell.text for cell in table.rows[0].cells]

        # Iterate over the remaining rows in the table and create a bullet point list
        bullet_points = []
        for row in table.rows[1:]:
            # Extract the values from each cell in the row
            values = [cell.text for cell in row.cells]
            # Create a string representation of the row as a bullet point
            bullet_point = '- ' + ', '.join([f'{header[i]}: {value}' for i, value in enumerate(values)])
            # Add the bullet point to the list of bullet points
            bullet_points.append(bullet_point)

        # Print the bullet point list for the current table
        print('Table:')
        print('\n'.join(bullet_points))
        print('\n')

#tables(document)
```
